```python
df = pd.read_csv("../../diaries_transactions_all.csv", low_memory=False);
df['trx_yr_mo'].value_counts()
```

### Only considering all formal loans

```python
frmloan = df[((df["trx_family_code"]=="FRMLN")&(df['trx_direction'] == 'Outflow'))]   
```

### Group formal loans by IDs

```python
frmloan.groupby(['m_ids_owner', 'trx_yr_mo']).groups.keys()
```

### Get the sum of the loan payments per id

```python
frmloan.groupby('m_ids_owner')['trx_value_kes'].sum()
```

### Montly paymetns amount 

```python
frmloan.groupby(['m_ids_owner', 'trx_value_kes'])['trx_yr_mo'].sum()
```

```python
frmloan.groupby(['trx_yr_mo','m_ids_owner']).agg({"trx_value_kes": "sum"})
```
