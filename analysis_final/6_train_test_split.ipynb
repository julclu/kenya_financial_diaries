```python
import pandas as pd
import glob
import os
from sklearn.model_selection import train_test_split
```

```python
glob.glob('*.csv')
```

```python
data = pd.read_csv('data_merged_noMissingData.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
data = data.loc[data['binarize_score'] != 999]
```

```python
data.binarize_score.value_counts()
```

```python
data.shape
```

```python
data.head()
```

```python
labels = data.binarize_score 
data = data.drop(['binarize_score'], axis = 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, labels, stratify = labels, random_state=42, test_size = 0.2 )
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
y_train = pd.DataFrame(y_train)
y_train.columns = ['owner_score_bin']
```

```python
y_test = pd.DataFrame(y_test)
y_test.columns = ['owner_score_bin']
```

```python
x_train.to_csv('binary_x_train.csv', index =0)
x_test.to_csv('binary_x_test.csv', index =0)
y_train.to_csv('binary_y_train.csv', index = 0)
y_test.to_csv('binary_y_test.csv', index = 0)
```

```python
train_df = pd.concat([x_train.reset_index(drop=True), y_train], axis=1)
test_df  = pd.concat([x_test.reset_index(drop=True), y_test], axis=1)
```

```python
train_df.to_csv('binary_train.csv', index =0)
test_df.to_csv('binary_test.csv', index = 0)
```
