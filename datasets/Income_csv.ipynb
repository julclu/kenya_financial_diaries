### The dominant income seems to be off. I will fix this now.

```python
for index,row in inc_new.iterrows():
    #number of income sources
    c = 0
    #we save all sources the individual has
    sources_ind = set()
    for source in inc_sources:
        if not math.isnan(inc_new.loc[index,"inc_i_meanpti_"+source]):
            if (inc_new.loc[index,"inc_i_meanpti_"+source] !=0) or ((inc_new.loc[index,"inc_i_sdpti_"+source] != 0) and not (math.isnan(inc_new.loc[index,"inc_i_sdpti_"+source]))):
                    c +=1
                    sources_ind.add(source)
                    
            
    inc_new.at[index,"inc_i_unique_sources"]=c
    
    #dominant income source
    if len(sources_ind)==0:
        continue
    if len(sources_ind)==1:
        inc_new.at[index,"inc_i_dom"]=list(sources_ind)[0]
    else:
        mean_source = []
        for source in sources_ind:
            mean_source.append("inc_i_meanpti_"+source)
        
        maxi = inc_new[inc_new["m_ids_owner"]==row["m_ids_owner"]][mean_source].idxmax(axis=1)
        dom = maxi.values[0]

        inc_new.at[index,"inc_i_dom"] = dom[14:]
```
