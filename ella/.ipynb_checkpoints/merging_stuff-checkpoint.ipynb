```python
demo = pd.read_csv("aux_data/demo_ind.csv",dtype={"m_ids":str})
```

```python
demo.head()
```

```python
inc = pd.read_csv("aux_data/inc_ind_over_trx_yr_mo_usd_excRR.csv",dtype={"m_ids_owner":str})
```

```python
inc.head()
```

```python
ids_dem = set(demo["m_ids"].unique())
```

```python
len(ids_dem)
```

```python
ids_inc = set(inc["m_ids_owner"].unique())
```

```python
len(ids_inc)
```

## check if there are matches

```python
if (ids_inc & ids_dem):
    print(len(ids_inc & ids_dem))
else:
    print("no matches")
```

## check for which ones there are no matches

```python
for ids in ids_inc:
    if ids in ids_dem:
        continue
    else: print(ids)
```
